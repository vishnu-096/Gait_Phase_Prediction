```python
import openpyxl
import os
import sys

import pandas as pd
import numpy as np
import math
from os import listdir, walk
from os.path import isfile, join
```

```python
def encode_phase(df, name):
    phase = df[name].values.tolist()
    x = np.zeros((len(df[name]),1))
    y = np.zeros((len(df[name]),1))
    for i in range (len(phase)): 
        phi = math.radians(phase[i])
        x[i] = math.cos(phi)
        y[i] = math.sin(phi)

    name_x = name[0] + 'hip_x'
    name_y = name[0] + 'hip_y'

    df[name_x] = x
    df[name_y] = y
```

```python
working_dir = os.getcwd()
sys_platform = sys.platform
excel_path_dict = {'linux' : "/csvs/", 'win32' : "\Data\Incline\\non-normal\\all subject\\"}
excel_path_name = excel_path_dict[sys_platform]

directory_path = working_dir
excel_path = directory_path + excel_path_name
```

```python
excel_path
```

```python
all_file = []

for(dirpath, dirnames, filenames) in walk(excel_path):
    all_file.extend(filenames)
  
```

```python
all_file
```

```python
for file in all_file:
    source_table = pd.read_excel(excel_path + file, sheet_name='Sheet1')
    source_table.info()
    encode_phase(source_table, 'l_ph_hip')
    encode_phase(source_table, 'r_ph_hip')
    source_table.info()
    source_table.to_excel(excel_path + file, sheet_name="Sheet1", index = False)
```
